# Extracting Segments

## K-Means Clustering Algorithm

```python
X=np.array(df1)
X
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans=kmeans.predict(X)
print(y_kmeans)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X[:,11],X[:,12],c=y_kmeans)
centers=kmeans.cluster_centers_
plt.scatter(centers[:,11],centers[:,12],c='red')
```

```python
# Data Distribution on graph 
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(40,40))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor='midnightblue')
        ax.set_title(feature+" Distribution",color='DarkRed')
        ax.set_yscale('log')
    fig.tight_layout()  
    plt.show()

draw_histograms(df1,df1.columns,8,4)
```

## Calulating target segments

```python
data['Clusters']=kmeans.labels_
data
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
data['Age'] = le.fit_transform(data['Age'])
age_freq = data.groupby('Clusters')['Age'].mean()
age_freq = age_freq.to_frame().reset_index()
age_freq
```

```python
data['VisitFrequency'] = le.fit_transform(data['VisitFrequency'])
visit_freq = data.groupby('Clusters')['VisitFrequency'].mean()
visit_freq = visit_freq.to_frame().reset_index()
visit_freq
```

```python
segment = age_freq.merge(visit_freq, on='Clusters', how='left')
plt.figure(figsize = (5,4))
sb.scatterplot(x = "Age", y = "VisitFrequency",data=segment,s=400, color="red")
plt.title("Simple segmentplot for this Mcdonalds dataset") 
plt.grid(True)
plt.xlabel("visit") 
plt.ylabel("Age") 
plt.show()
```
